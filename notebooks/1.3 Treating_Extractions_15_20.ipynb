```python
import pandas as pd
import numpy as np
import re
import pymupdf

from PyPDF2 import PdfReader
```

```python
df_2015_2017 = pd.read_csv('../data/extracted_questions/df_15-17.csv')

df_2015_2017.head()
```

```python
df_2018_2020 = pd.read_csv('../data/extracted_questions/df_18-20.csv')

df_2018_2020.head()
```

```python
df_2015_2017 = df_2015_2017.rename(columns={
    'CO_POSICAO': 'numero_questao',
    'QUESTOES': 'enunciado',
    'ALTERNATIVAS': 'alternativas',
    'NU_PARAM_B':'nu_param_B',
    'TX_GABARITO':'gabarito'
})

df_2018_2020 = df_2018_2020.rename(columns={
    'CO_POSICAO': 'numero_questao',
    'QUESTOES': 'enunciado',
    'ALTERNATIVAS': 'alternativas',
    'NU_PARAM_B':'nu_param_B',
    'TX_GABARITO':'gabarito'
})
```

```python
df_2015_2017.drop(columns = ['Unnamed: 0','NU_PARAM_A','NU_PARAM_C'], inplace = True)

df_2018_2020.drop(columns = ['NU_PARAM_A','NU_PARAM_C'], inplace = True)
```

```python
df_2015 = df_2015_2017[df_2015_2017['ANO'] == 2015]

df_2016 = df_2015_2017[df_2015_2017['ANO'] == 2016]

df_2017 = df_2015_2017[df_2015_2017['ANO'] == 2017]

df_2018 = df_2018_2020[df_2018_2020['ANO'] == 2018]

df_2019 = df_2018_2020[df_2018_2020['ANO'] == 2019]

df_2020 = df_2018_2020[df_2018_2020['ANO'] == 2020]
```

```python
dfs_by_year = {
    2015: df_2015,
    2016: df_2016,
    2017: df_2017,
    2018: df_2018,
    2019: df_2019,
    2020: df_2020,
}
```
